```python
def PostList(file):
    if file in os.listdir():
        ff = open(file,'r')
        s = list(ff.read().split())

        post = {}

        n = len(s)
        
        for i in range(n):
            if s[i] in post:
                post[s[i]][1].append(i)
            else :
                post[s[i]] = [file,[i]]

        return post
    
    
    else : return '#'
```
