```python
#Query 
import os

folder='Z:\SEM5\IR-LAB\Documents'
os.chdir(folder)

print(os.listdir())

diction = {}

special='\'!.,:*?[]()'

for j in os.listdir():
    ff = open(j,'r')
    for line in ff:
        for zz in special:
            line = line.replace(zz, ' ')
        ll = list(set(line.lower().split()))
        for w in ll:
            if w in diction:
                diction[w].append(j)
            else :
                diction[w] = [j]

for i,j in diction.items():
    diction[i] = list(set(j))

for i,j in diction.items():
    print(i,'::',j)
        
```

```python
class stack:
    
    def __init__(self):
        self.cont = []
        self.top = -1
    
    def push(self,val):
        self.cont.append(val)
        self.top+=1
    
    def pop(self):
        if(self.top!=-1):
            self.top-=1
            return self.cont.pop()
    def isemp(self):
        return self.top==-1
    
    def count(self):
        return self.top+1

def isop(val):
    oper = ['and','or','not']
    if val in oper:
        return True
    else : return False
    
def prec(op):
    prc = {
        'or':1,
        'and':2,
        'not':3
    }
    return prc[op]
```

```python
def conv(st):
    ex = st.lower().split()
    s = stack()
    exp = []
    
    for i in ex:  
        if isop(i):
            while (not s.isemp()) and  prec(s.cont[-1])>=prec(i):
                exp.append(s.pop())
            s.push(i)
        else:
            exp.append(i)

    while not s.isemp():
        exp.append(s.pop())
    
    return exp
```

```python
def getPost(val):
    val = val.lower().strip()
    if val in diction:
        return diction[val]
    else : return '#'
    
```

```python
def getPost(a):
    return diction[a]

def intersec(a,b):
        res = []
        for i in a:
            if i in b:
                res.append(i)
        return res

def union(a,b):
    return list(set(list(a)+list(b)))
```

```python
def search(st):
    filesall = set(os.listdir())
    exp = conv(st)
#     print(exp)
    l = stack()
    res = {}
    
    for i in exp:
#         print(l.cont)
        if not isop(i):
            l.push(i)
        elif isop(i):
#             op1 = l.pop()
#             op2 = l.pop()
            if i == 'and':
                op1 = l.pop()
                op2 = l.pop()              
                res=intersec(getPost(op1),getPost(op2))
                diction['resp'] = res
                l.push('resp')
            elif i == 'or':
                op1 = l.pop()
                op2 = l.pop()
                res= union(getPost(op1),getPost(op2))
                diction['resp'] = res
                l.push('resp')
            elif i == "not":
                op1 = l.pop()
                ans = filesall.difference(set(getPost(op1)))
                diction['not'+op1] = ans
                l.push('not'+op1)
                print(l.cont)
                print(diction[l.cont[-1]])

    return diction[l.cont[-1]]
        

    
```
